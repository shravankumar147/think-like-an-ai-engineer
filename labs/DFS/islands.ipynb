```python
grid = [
    [1,1,0],
    [1,0,0],
    [0,0,1]
]

rows, cols = len(grid), len(grid[0])

def print_grid():
    for row in grid:
        print(row)
    print()

print("🌍 Initial grid:")
print_grid()

num_islands = 0

for r in range(rows):
    for c in range(cols):
        if grid[r][c] == 1:
            print(f"🚀 Found NEW island starting at ({r}, {c})")
            num_islands += 1

            stack = [(r, c)]

            while stack:
                print(f"\n📦 Current stack: {stack}")
                row, col = stack.pop()
                print(f"👉 Popped cell: ({row}, {col})")

                # Step 1: boundary check
                if row < 0 or row >= rows or col < 0 or col >= cols:
                    print("❌ Out of bounds → skip")
                    continue

                # Step 2: skip water or visited
                if grid[row][col] == 0:
                    print("🌊 Water or already visited → skip")
                    continue

                # Step 3: mark visited
                print(f"✅ Visiting land at ({row}, {col}) → marking as 0")
                grid[row][col] = 0

                print("🗺️ Grid after marking:")
                print_grid()

                # Step 4: explore neighbors
                print("🔄 Adding neighbors (up, down, left, right)")
                stack.append((row - 1, col))  # up
                stack.append((row + 1, col))  # down
                stack.append((row, col - 1))  # left
                stack.append((row, col + 1))  # right

            print(f"🏝️ Finished exploring island starting at ({r}, {c})\n")

print("🏁 Final grid:")
print_grid()

print(f"🌟 Total islands found: {num_islands}")
```

```python
grid = [
    [1,1,0],
    [1,0,0],
    [0,0,1]
]

rows, cols = len(grid), len(grid[0])

def print_grid():
    for row in grid:
        print(row)
    print()

def dfs(r, c, depth=0):
    indent = "  " * depth  # helps visualize recursion depth

    print(f"{indent}👉 Enter dfs({r}, {c})")

    # Step 1: boundary check
    if r < 0 or r >= rows or c < 0 or c >= cols:
        print(f"{indent}❌ Out of bounds → return")
        return

    # Step 2: skip water or visited
    if grid[r][c] == 0:
        print(f"{indent}🌊 Water/visited at ({r}, {c}) → return")
        return

    # Step 3: mark visited
    print(f"{indent}✅ Visiting land at ({r}, {c}) → marking 0")
    grid[r][c] = 0

    print(f"{indent}🗺️ Grid now:")
    print_grid()

    # Step 4: explore neighbors
    print(f"{indent}🔄 Explore neighbors of ({r}, {c})")

    dfs(r-1, c, depth+1)  # up
    dfs(r+1, c, depth+1)  # down
    dfs(r, c-1, depth+1)  # left
    dfs(r, c+1, depth+1)  # right

    print(f"{indent}⬅️ Return from dfs({r}, {c})")


print("🌍 Initial grid:")
print_grid()

num_islands = 0

for r in range(rows):
    for c in range(cols):
        if grid[r][c] == 1:
            print(f"\n🚀 Found NEW island at ({r}, {c})")
            num_islands += 1
            dfs(r, c)

print("\n🏁 Final grid:")
print_grid()

print(f"🌟 Total islands found: {num_islands}")
```
